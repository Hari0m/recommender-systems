```python
moviemat= df.pivot_table(index='user_id',columns='title',values='rating')
```

```python
moviemat.head()
```

```python
ratings.sort_values('Num of Ratings',ascending=False)
```

```python
starwars = moviemat['Star Wars (1977)']
liar=moviemat['Liar Liar (1997)']
```

```python
moviemat.corrwith(starwars)
```

```python
similarstar=moviemat.corrwith(starwars)
```

```python
similarliar=moviemat.corrwith(liar)
```

```python
corrstar=pd.DataFrame(similarstar,columns=['Correlation'])
corrstar.dropna(inplace=True)
```

```python
corrstar.head()
```

```python
corrstar=corrstar.join(ratings['Num of Ratings'])
```

```python
corrstar.head()
```

```python
corrstar[corrstar['Num of Ratings']>100].sort_values('Correlation',ascending=False)
```

```python
corrliar = pd.DataFrame(similarliar,columns=['Correlation'])
```

```python
corrliar.dropna(inplace=True)
```

```python
corrliar=corrliar.join(ratings['Num of Ratings'])
```

```python
corrliar[corrliar['Num of Ratings']>100].sort_values('Correlation',ascending=False)
```
